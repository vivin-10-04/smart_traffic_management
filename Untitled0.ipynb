```python
import csv
import os
def read_specific_column(file_path, column_name):
    with open(file_path, mode='r') as file:
        reader = csv.DictReader(file)

        # Ensure the column exists
        if column_name not in reader.fieldnames:
            raise ValueError(f"Column '{column_name}' not found in the file.")

        # Extract data from the specified column
        column_data = [row[column_name] for row in reader]

    return column_data

def get_latest_directory(parent_dir):
    # List all subdirectories in the parent directory
    subdirs = [os.path.join(parent_dir, d) for d in os.listdir(parent_dir) if os.path.isdir(os.path.join(parent_dir, d))]

    # Ensure there are subdirectories
    if not subdirs:
        return None  # No subdirectories found

    # Find the directory with the latest modification time
    latest_dir = max(subdirs, key=os.path.getmtime)
    return latest_dir

# Example usage
parent_directory = "/content/drive/MyDrive/paper/runs/detect"
latest_directory = get_latest_directory(parent_directory)
print(latest_directory)
file_path = latest_directory + "/results.csv"
column_name = "metrics/recall(B)"
column_data = read_specific_column(file_path, column_name)
print(column_data)
```

```python
def emergency_vehicle_observed(column_data):
    for value in column_data:
        if float(value) >= 0.5:
            return 1
    return 0

emergency_vehicle_observed(column_data)
```

```python
class TrafficLight:
    def __init__(self, id):
        self.id = id  # Unique identifier for each traffic light
        self.state = 'red'  # Initial state is red

    def switch_to_green(self):
        self.state = 'green'
        print(f"Traffic Light {self.id} is now Green.")

    def switch_to_red(self):
        self.state = 'red'
        print(f"Traffic Light {self.id} is now Red.")
```

```python
class SDNController:
    def __init__(self, traffic_lights):
        self.traffic_lights = traffic_lights  # List of TrafficLight objects

    def switch_traffic_light(self, traffic_light_id, signal):
        # Ensure only one light is green at a time
        if signal == 1:  # Switch the selected traffic light to Green
            for light in self.traffic_lights:
                if light.id != traffic_light_id:
                    light.switch_to_red()  # Turn all other lights red
            self.traffic_lights[traffic_light_id - 1].switch_to_green()  # Switch the selected light to Green
        else:  # Signal is 0, turn the selected traffic light to Red
            self.traffic_lights[traffic_light_id - 1].switch_to_red()
```

```python
traffic_lights = [TrafficLight(id=i) for i in range(1, 4)]  # Example: 3 traffic lights for every intersction

# Create SDN Controller
sdn_controller = SDNController(traffic_lights)


inputs = [(1,emergency_vehicle_observed(column_data)),(2,0),(3,0)]

# Process inputs to switch traffic lights
for light_id, signal in inputs:
    sdn_controller.switch_traffic_light(light_id, signal)
```
